```python
from pyspark.sql.functions import col, upper
customers = spark.read.table("02_silver_catalog.default.customers")
```

```python
from pyspark.sql.functions import regexp_replace, initcap, upper, col, to_date

customers = customers.withColumn("country", regexp_replace(col("country"), "\.", "")) \
    .withColumn("country", upper(col("country"))) \
    .withColumn("customer_type", initcap(col("customer_type"))) \
    .withColumn("signup_date", col("signup_date").cast("date"))

customers.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("02_silver_catalog.default.customers")
```
