```python
from pyspark.sql.functions import when, col, lit

invoice_line_items_df = spark.table("02_silver_catalog.transformed_tables.invoice_line_items")
invoice_line_items_df = invoice_line_items_df.withColumn(
    "quantity",
    when(col("quantity") < 0, lit(0)).otherwise(col("quantity"))
)
invoice_line_items_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("02_silver_catalog.transformed_tables.invoice_line_items")
# display(invoice_line_items_df)
```
