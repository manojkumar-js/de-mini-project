```python
from pyspark.sql.functions import upper, initcap

invoices_df = spark.table("02_silver_catalog.default.invoices")
invoices_transformed = invoices_df.withColumn("region", upper("region")) \
                                 .withColumn("invoice_status", initcap("invoice_status"))
display(invoices_transformed)

invoices_transformed.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("02_silver_catalog.default.invoices")
```
