```python
tables = [
    "customers",
    "exchange_rates",
    "invoice_line_items",
    "invoices",
    "products",
    "payments"
]

for table in tables:
    df = spark.read.table(f"02_silver_catalog.default.{table}")
    df_clean = df.dropDuplicates()
    df_clean.write.mode("overwrite").saveAsTable(f"02_silver_catalog.default.{table}")
```
