```python
%sql
    
CREATE OR REPLACE TABLE 03_gold_catalog.analytical_tables.cube_sales_analytics AS
SELECT
    fs.invoice_line_id,
    fs.invoice_id,
    fs.invoice_date,
    fs.invoice_status,
    fs.quantity,
    fs.unit_price,
    fs.discount,
    fs.gross_amount,
    fs.discount_amount,
    fs.net_amount_local,
    fs.revenue_usd,
    fs.cost_total,
    fs.profit_local,
    fs.profit_usd,
    fs.currency,
    fs.rate_to_usd,

    dc.customer_id,
    dc.customer_name,
    dc.customer_type,
    dc.country AS customer_country,
    dc.signup_date AS customer_signup_date,
    
    dp.product_id,
    dp.product_name,
    dp.category AS product_category,
    dp.cost_price,
    
    fs.region,
    dr.country AS region_country,
    
    YEAR(fs.invoice_date) AS invoice_year,
    MONTH(fs.invoice_date) AS invoice_month,
    QUARTER(fs.invoice_date) AS invoice_quarter,
    DAYOFWEEK(fs.invoice_date) AS invoice_day_of_week,
    DATE_FORMAT(fs.invoice_date, 'MMMM') AS invoice_month_name,
    DATE_FORMAT(fs.invoice_date, 'EEEE') AS invoice_day_name
    
FROM 03_gold_catalog.analytical_tables.fact_sales fs
LEFT JOIN 03_gold_catalog.analytical_tables.dim_customers dc
    ON fs.customer_id = dc.customer_id
LEFT JOIN 03_gold_catalog.analytical_tables.dim_products dp
    ON fs.product_id = dp.product_id
LEFT JOIN 03_gold_catalog.analytical_tables.dim_regions dr
    ON fs.region = dr.region_name;
```
