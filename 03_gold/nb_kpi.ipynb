```python
%sql
-- KPI 1
SELECT ROUND(SUM(revenue_usd), 2) AS total_revenue_usd FROM 03_gold_catalog.analytical_tables.cube_sales_analytics WHERE invoice_status != 'Cancelled';
```

```python
%sql
-- KPI 2
SELECT
    SUM(quantity) AS total_quantity_sold
FROM 03_gold_catalog.analytical_tables.cube_sales_analytics
WHERE invoice_status != 'Cancelled';
```

```python
%sql
-- KPI 3
SELECT
    ROUND(AVG(invoice_total_usd), 2) AS avg_invoice_value_usd
FROM (
    SELECT
        invoice_id,
        SUM(revenue_usd) AS invoice_total_usd
    FROM 03_gold_catalog.analytical_tables.cube_sales_analytics
    WHERE invoice_status != 'Cancelled'
    GROUP BY invoice_id
);
```

```python
%sql
-- KPI 4
SELECT
    customer_id,
    customer_name,
    customer_type,
    customer_country,
    ROUND(SUM(revenue_usd), 2) AS total_revenue_usd,
    COUNT(DISTINCT invoice_id) AS num_invoices
FROM 03_gold_catalog.analytical_tables.cube_sales_analytics
WHERE invoice_status != 'Cancelled'
GROUP BY customer_id, customer_name, customer_type, customer_country
ORDER BY total_revenue_usd DESC
LIMIT 10;
```

```python
%sql
-- KPI 5
SELECT
    product_id,
    product_name,
    product_category,
    ROUND(SUM(revenue_usd), 2) AS total_revenue_usd,
    SUM(quantity) AS total_quantity_sold
FROM 03_gold_catalog.analytical_tables.cube_sales_analytics
WHERE invoice_status != 'Cancelled'
GROUP BY product_id, product_name, product_category
ORDER BY total_revenue_usd DESC
LIMIT 10;
```

```python
%sql
-- KPI 6
SELECT
    COUNT(DISTINCT CASE WHEN invoice_status = 'Cancelled' THEN invoice_id END) AS cancelled_invoices,
    COUNT(DISTINCT invoice_id) AS total_invoices,
    ROUND(
        (COUNT(DISTINCT CASE WHEN invoice_status = 'Cancelled' THEN invoice_id END) * 100.0) /
        NULLIF(COUNT(DISTINCT invoice_id), 0),
        2
    ) AS cancellation_rate_percent
FROM 03_gold_catalog.analytical_tables.cube_sales_analytics;
```

```python
%sql
-- KPI 7
SELECT
    ROUND(SUM(discount_amount), 2) AS total_discount_given_local,
    ROUND(SUM(discount_amount * rate_to_usd), 2) AS total_discount_given_usd,
    ROUND(AVG(discount * 100), 2) AS avg_discount_percent,
    COUNT(DISTINCT customer_id) AS customers_receiving_discount
FROM 03_gold_catalog.analytical_tables.cube_sales_analytics
WHERE invoice_status != 'Cancelled'
    AND discount > 0;
```

```python
%sql
    
-- KPI 8
SELECT
    region,
    ROUND(SUM(revenue_usd), 2) AS total_revenue_usd,
    COUNT(DISTINCT invoice_id) AS num_invoices,
    COUNT(DISTINCT customer_id) AS num_customers,
    ROUND(AVG(revenue_usd), 2) AS avg_revenue_per_transaction_usd
FROM 03_gold_catalog.analytical_tables.cube_sales_analytics
WHERE invoice_status != 'Cancelled'
GROUP BY region
ORDER BY total_revenue_usd DESC;
```

```python
%sql
-- KPI 9
SELECT
    customer_id,
    customer_name,
    customer_type,
    COUNT(DISTINCT invoice_id) AS num_invoices,
    ROUND(SUM(revenue_usd), 2) AS total_revenue_usd,
    ROUND(AVG(revenue_usd), 2) AS avg_transaction_value_usd
FROM 03_gold_catalog.analytical_tables.cube_sales_analytics
WHERE invoice_status != 'Cancelled'
GROUP BY customer_id, customer_name, customer_type
ORDER BY num_invoices DESC, total_revenue_usd DESC
LIMIT 20;
```
